```python
class Lampada:
    def __init__(self, cor, voltagem, luminosidade):
        self.__cor = cor
        self.__voltagem = voltagem
        self.__luminosidade = luminosidade
        self.__ligada = 0
        
    def verificar_ligada(self):
        if self.__ligada == 0:
            return "Desligada"
        else:
            print("Ligada")
    
    def ligar_desligar(self):
        if self.__ligada == 0:
            self.__ligada = 1
            return "lampada ligada"
        if self.__ligada == 1:
            self.__ligada = 0
            return "lampada desligada"
        
lamp1=Lampada('azul',110,80)
print(lamp1.verificar_ligada())  #Inicializa desligada
print(lamp1.ligar_desligar())  #Ligada
print(lamp1.ligar_desligar())  #Desligada
print()

        
#outra forma de resolver


class Lampada:
    
    def __init__(self,cor,voltagem,luminosidade):
        self.__cor=cor
        self.__voltagem=voltagem
        self.__luminosidade=luminosidade
        self.__ligada=False
    
    def verificar_ligada(self):
        return self.__ligada
 
    def ligar_desligar(self):
        if self.__ligada:
            self.__ligada=False
            print("teste", self.__ligada)
            return self.__ligada
        else:
            self.__ligada=True
            return self.__ligada
            
lamp1=Lampada('azul',110,80)
print(lamp1.verificar_ligada())  #Inicializa desligada
print(lamp1.ligar_desligar())  #Ligada
print(lamp1.ligar_desligar())  #Desligada
```

<i><b> Crie uma classe Lampada com os seguintes atributos e métodos:

## Atributos:

### - Cor

### - Voltagem

### - Ligada (a lâmpada deve inicializar desligada)

## Métodos:

### - Verificar se a lâmpada está ligada/desligada;

### - Ligar/desligar a lâmpada

### - Todos os atributos devem ser privados.



## crie um objeto da classe Lampada e teste os métodos criados.
</b></i>     
</p>

---

```python
import math

class Vetor:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vetor(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vetor(self.x - other.x, self.y - other.y)
    
    def __mul__(self, other):
        return self.x * other.x + self.y * other.y
    
    def __abs__(self):
        return math.sqrt(self.x**2 + self.y**2)
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
    def __ne__(self, other):
        return not self.__eq__(other)
    
    def __str__(self):
        return f"({self.x}, {self.y})"

# Criar dois vetores para teste
vetor1 = Vetor(3, 4)
vetor2 = Vetor(1, 2)

# Testar os métodos implementados
print("Soma:", vetor1 + vetor2)
print("Subtração:", vetor1 - vetor2)
print("Produto interno:", vetor1 * vetor2)
print("Norma do vetor 1:", abs(vetor1))
print("Vetor 1 é igual ao vetor 2?", vetor1 == vetor2)
print("Vetor 1 é diferente do vetor 2?", vetor1 != vetor2)
```
